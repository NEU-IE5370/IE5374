```python
s = map(sum,zip(range(0,1_000_000),range(1_000_000,2_000_000)))
sys.getsizeof(s)
```

```python
timeit.timeit(lambda: list(map(sum,zip(range(0,1_000_000),range(1_000_000,2_000_000)))),number = 10)
```
