# Generators and Sequences

```python
def ln2Series():
    n=1
    series = 0
    term = 10000
    while (abs(term) > 0.001):
        a = (-1)**(n-1)
        term = a/n
        series += term
        n+=1
    print(n,":",series)
    return series

ln2Series()
print(math.log(2))
```

```python
def seriesGen(term):
    n=0
    while True:
        yield term(n)
        n+=1

def ln2Term(n):
    if n==0:return 0
    return ((-1)**(n-1))/n
```

```python
partial_sum = accumulate(seriesGen(ln2Term))
```

```python
list(islice(partial_sum,1002,1003))[0]
```

```python
print(math.log(2))
```

```python
ps = accumulate(map(ln2Term,count(1)))
list(islice(ps,1001,1002))[0]
```
