```python
 class point2D:    
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
        self.__callnumber = 0
        
    def __called(self):
        self.__callnumber = self.__callnumber+1
    
    def mag(self):
        self.__called()
        return sqrt(pt.x**2 + pt.y**2)
    
    def dot(self,pt):
        self.__called()
        return self.x*pt.x+self.y*pt.y
        
    def __add__(self,pt):
        self.__called()
        x = self.x+pt.x
        y = self.y+pt.y
        return point2D(x,y)
    
    def __mul__(self, val):
        'x*val'
        self.__called()
        if type(val) in (type(1),type(1.0)):
            x = self.x*val
            y = self.y*val 
            return point2D(x,y)
        else: raise TypeError
            
    def __cmp__(self,other):
        self.__called()
        pass
    
    def __rmul__(self, other):
        '2*x'
        return self.__mul__(other)

    
    def __str__(self):
        self.__called()
        return '[{},{}]'.format(self.x, self.y)
    
    def __del__(self):
        print( self.__callnumber)
```

## Generators Have State

```python
def summationCubeGenerator(n):
    total, k = 0, 1
    while k <= n:
        total, k = total + pow(k,3), k + 1
        print("state: [{},{},{}]".format(total,k,n))
        yield total
```

```python
s = summationCubeGenerator(10)
```

```python
next(s)
```

```python
next(s)
```

# What is the difference between it and this class?

```python
class cuber:
    def __init__(self,n):
        self.total = 0
        self.k = 1
        self.n = n

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.k<=self.n:
            self.total = self.total + pow(self.k,3)
            self.k = self.k + 1
            print("state: [{},{}]".format(self.total,self.k))
            return self.total
        raise StopIteration
    
    
c = cuber(10)
```

```python
next(c)
```

```python
next(c)
```

```python
from itertools import *

c = cuber(10)
list(islice(c,0,20))
```

```python
s = summationCubeGenerator(10)
list(islice(s,0,20))
```

## 1 Impure functions and side effects

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return "Name: {}, {} years old".format(self.name,
                                              self.age)
    def birthday(self):
        n = self.name.capitalize()
        print("HAPPY BIRTHDAY {}".format(n))
        self.age = self.age+1


```

```python
Methuselah= person("Methuselah",969)
print(Methuselah)

Methuselah.birthday()
print(Methuselah)
```

```python
class professor(person):
    def __init__(self,name,age,university):
        super().__init__(name,age)
        self.uni = university
        self.classes = []
        
    def addClass(self,className):
        self.classes.append(className)
    
    def classList(self):
        classesString=", ".join(self.classes)
        return "Teaches "+classesString+" at {}: ".format(self.uni)
    
```
